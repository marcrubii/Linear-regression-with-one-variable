```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# We introduce our data to do the representation
x_train = np.array([10,20,50,70,85]) # Our input x
y_train = np.array([15.0,20.0,25.0,35.0,45.0]) # Our output y
m = len(x_train) # m is our number of training examples
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
print(f"number of training examples is: {m}")
```

```python
# We plot the data points to see our data
plt.scatter(x_train, y_train, marker = 'x' , c = 'r')
plt.title ("House prices")
plt.ylabel("Price of the houses (in thousands)")
plt.xlabel("size of the house (m²)")
plt.figure(figsize=(1, 2))
plt.show
```

```python
# We initialize the parametres w (weight) and b (bias)
w = 1
b = 0.15
def prediction(x, w, b):
    m = x.shape[0]
    f_wb = np.zeros(m) # We create an array of m length
    for i in range (m):
        f_wb[i] = w * x[i] + b #Linear regression function with one variable
    
    return f_wb
```

```python
# Now, we can plot our prediction
test_f_wb = prediction(x_train, w, b)
plt.plot(x_train, test_f_wb, c = 'b', label = 'Our prediction with the model')
# And plot our data 
plt.scatter(x_train, y_train, marker ='x', c='r', label='Actual values of the data')
plt.title("Housing prices prediction")
plt.ylabel('Price of the houses (in thousands)')
plt.xlabel("size of the house (m²)")
plt.legend()
plt.show()
```
